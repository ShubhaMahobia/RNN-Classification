```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
word_index = imdb.get_word_index()
reverse_word_index = {value: key for key, value in word_index.items()}
```

```python
model = load_model('simple_rnn_imbd.h5')
```

```python
model.get_weights()
```

```python
##Helper Function - 
def pre_process_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word, 2) + 3 for word in words]
    padded_review = sequence.pad_sequences([encoded_review], maxlen = 500)
    return padded_review
```

```python
def predict_sentiment(review):
    preProccessed_text = pre_process_text(review)
    prediction = model.predict(preProccessed_text)
    sentiment = 'Postive' if prediction[0][0] > 0.5 else 'Negative'
    return sentiment , prediction[0][0]
```

```python
sample_example = "The Movie was simple"
sentiment, score = predict_sentiment(sample_example)
print(f'Review {sample_example}')
print(f'Sentiment {sentiment}')
print(f'Prediction Score {score}')
```
